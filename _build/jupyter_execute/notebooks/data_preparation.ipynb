```python
from sklearn.preprocessing import OrdinalEncoder
textual_columns = ['Month', 'VisitorType', 'Weekend', 'Revenue']
enc = OrdinalEncoder()
df[textual_columns] = enc.fit_transform(df[textual_columns])

df.head()
```
